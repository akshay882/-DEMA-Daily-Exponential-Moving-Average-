```python
# DEMA (Daily Expotential Moving Average) with Python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_name = 'NIFTY 50_Data.csv'

df1 = pd.read_csv(file_name, index_col='Date')

# EMA calculation :

df1['EMA200'] = df1['Close'].ewm(span=200).mean()
df1['EMA50'] = df1['Close'].ewm(span=50).mean()

print(df1.head(3))

# PLOT

sns.set(rc={'figure.figsize':(20,10)})
custom_params = {"axes.spines.bottom": False,"axes.spines.top": False,"axes.spines.right": False}
sns.set_theme(style="white", rc = custom_params)

sns.lineplot(data=df1[['Close','EMA200','EMA50']], 
            palette=['blue', 'red', 'green'], markers=False, dashes=False)
```
